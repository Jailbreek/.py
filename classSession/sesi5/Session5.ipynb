```python
# Import library for PDB

from Bio.PDB.PDBParser import  PDBParser

parser = PDBParser()
```

```python
structure_pdb = parser.get_structure('6lu7', './6lu7.pdb')

structure_pdb
```

```python
model = structure_pdb[0]
model
```

```python
# print to check chain in model

for chain in model:
    print(chain)
    for residue in chain:
        print(residue)
        for atom in residue:
            print(atom)
```

```python
from Bio import SeqIO
dir(SeqIO)
```

```python
fasta_record = SeqIO.parse('./coronavirus.fasta', 'fasta')
for record in fasta_record:
    print(record.id)
    print(record.description)
    print(record.seq)
    for protein in record.seq.translate().split('*'):
        print(protein)
        print(len(protein))
    # print(record.seq.translate().split('*'))
```
